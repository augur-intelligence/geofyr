```python
import pandas as pd
```

```python
wiki_dat = pd.read_parquet('data/wiki_geo_data.parquet')
```

```python
names_dat = pd.read_csv('data/geonames_data.csv')
```

```python
df_union = pd.concat([wiki_dat, names_dat])
```

```python
df_union.sample(frac=0.2).plot.scatter(x='lon', y='lat', figsize = (45,35), s=0.01)
```

```python
df_union.to_parquet("data/geo_data.parquet")
# df_union = pd.read_parquet("data/geo_data.parquet")
```
